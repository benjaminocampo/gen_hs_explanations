```python
def parse_string(input_string):
    lines = input_string.strip().split("\n")
    parsed_data = {}

    for line in lines:
        key, value = line.split(": ", 1)
        parsed_data[key] = value

    return parsed_data
```

```python
import re

def split_string(input_string):
    try:
        # Define the regular expression pattern
        # The pattern now includes a non-greedy match for any character following the labels
        pattern = r'(Label:.*?|Target:.*?|Explanation:.*?)($|(?= Label:| Target:| Explanation:))'
        # Use findall to find all matches of the pattern in the input string
        matches = re.findall(pattern, input_string)

        # Return the list of matches, removing the empty groups from the result
        lines = [match[0] for match in matches]

        parsed_data = {}
        for line in lines:
            key, value = line.split(":", 1)
            parsed_data[key.strip()] = value.strip()

        return parsed_data
    except:
        return None

# Example usage
input_str = 'Label: hateful Target: women Explanation: The message is hateful because it implies women are inferior.'
result = split_string(input_str)
print(result)
```

```python

hatecheck_t5_small["parsed_gen_output"] = hatecheck_t5_small["pred_exp"].apply(split_string)
hatecheck_t5_small["pred_labels"] = hatecheck_t5_small["parsed_gen_output"].apply(lambda r: r.get("Label"))
hatecheck_t5_small["pred_targets"] = hatecheck_t5_small["parsed_gen_output"].apply(lambda r: r.get("Target"))
hatecheck_t5_small["pred_exps"] = hatecheck_t5_small["parsed_gen_output"].apply(lambda r: r.get("Explanation"))
```
